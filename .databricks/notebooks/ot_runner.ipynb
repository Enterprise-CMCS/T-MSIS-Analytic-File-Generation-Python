```
state_code = dbutils.widgets.get("state_code")
run_id = dbutils.widgets.get("run_id")
```

```
if len(state_code) and len(run_id):
  if (state_code.find(',') != -1) and (run_id.find(',') != -1):
    state_list = list(eval(state_code))
    run_id_list = list(eval(run_id))
  else:
    state_list = [eval(state_code)]
    run_id_list = [eval(run_id)]
else:
  state_list = []
  run_id_list = []
```

```
from taf.OT.OT_Runner import OT_Runner
```

```
from taf.OT.OT_Metadata import OT_Metadata
```

```
acc = 0

for s in state_list:
  state = f"""'{ s }'"""
  run_id = f"""'{ run_id_list[acc] }'"""
  
  otr = OT_Runner(reporting_period=dbutils.widgets.get("reporting_period")
                 ,state_code=state
                 ,run_id=run_id)

  otr.print()
  otr.init()
  otr.run()
  otr.view_plan()
  display(otr.audit())
  
  acc += 1
```

```
logger = logging.getLogger("taf_log")
```

```
logger.setLevel(logging.DEBUG)
```

```
reporting_period = dbutils.widgets.get("reporting_period")
```

```
state_code = dbutils.widgets.get("state_code")
```

```
run_id = dbutils.widgets.get("run_id")
```

```
if len(state_code) and len(run_id):
  if (state_code.find(',') != -1) and (run_id.find(',') != -1):
    state_list = list(eval(state_code))
    run_id_list = list(eval(run_id))
  else:
    state_list = [eval(state_code)]
    run_id_list = [eval(run_id)]
else:
  state_list = []
  run_id_list = []
```

```
domain = f"{dbutils.secrets.get(scope='taf', key='DAPI_DOMAIN')}"
```

```
token_1 = "token:" + f"{dbutils.secrets.get(scope='taf', key='DAPI_TOKEN')}"
```

```
headers = {"Authorization": "Basic " + (base64.standard_b64encode(token_1.encode('utf-8'))).decode()}
```

```
headers["Accept"] = "application/json"
```

```
acc = 0

for s in state_list:
  state = f"'{ s }'"
  run_id = f"'{ run_id_list[acc] }'"
  
  json_params = {"job_id": 631594, "notebook_params": {"reporting_period": reporting_period, "state_code": state, "run_id": run_id}}
  
  try:
    response = requests.post('https://%s/api/2.0/jobs/run-now' % (domain), json=json_params, verify=False, headers=headers)
    response.raise_for_status()
    response_data = response.json()
    logger.info(f"Response Text: {response_data}")
    logger.info(f"Response: {response}")
  except Exception as err:
    logger.error(f"A error {err} was encountered!")
  else:
    print(f"Running: {json_params}")
    
  acc += 1
```
